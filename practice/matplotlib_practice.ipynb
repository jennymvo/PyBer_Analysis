```python
#import dependencies

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
#create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Other way of writing the code above:
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
#a third way of writing object oriented language with less lines
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
#Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker='*', color='blue', linewidth=2, label = 'Boston')

#Create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y Limit between 0 and 45
plt.ylim(0,45)

#Create a title
plt.title('PyBer Fare by month')

#Add a grid
plt.grid()

#Add the legend
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create a bar chart using the object oriented interface method
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create a horizontal bar chart using the object oriented method
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
#switch the data on both sides
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```

```python
#Create a scatterplot using the plot() function
plt.plot(x_axis, y_axis, 'o')
```

```python
#Create a scatterplot using the scatter() function
plt.scatter(x_axis, y_axis)
```

```python
#Create a bubble chart

#Change the size of the marker in the graph with "s" = size perameter
#The size of each marker is equal to the data floating point decimal values of the y-axis
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#Create a scatter plot with larger bubbles

#adjust the size of the bubbles by multiplying the data. 
#multiply the data by 3
y_axis_larger =[]
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#Create a scatter plot using the list comprehension technique
#format: new_list = [expression for item in list if conditional]

plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
#Create a scatterplot using the object oriented interface

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
#Create a bubblechart using the object oriented interface
fig, ax = plt.subplots()
ax = plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#Create a pie chart using the MATLAB method
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
#Make the largest percentage explode
#explode values indicate offset by the fraction of the radius using an array-like data points for each value in the chart. 
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
#a percentage of each wedge is added using autopct with one dicimal place using .1f%. the % % formats the number as percentage  
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
#change the wedge colors

# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
